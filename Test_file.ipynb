```python
def can_predate(predator, prey):
    """
    predator: a row/object of the hunting species
    prey: a row/object of the potential victim
    """
    # 1. Non-carnivores don't hunt other birds in our simple model
    if predator['Diet-5Cat'] not in ('Carnivore','Omnivore'):
        return False
    
    # 2. Size Constraint: A bird usually needs to be significantly 
    # larger than its prey to kill it.
    # We'll use a 1.5x mass ratio as a baseline.
    if predator['BodyMass-Value'] > (prey['BodyMass-Value'] * 1.5):
        return True
    
    return False
```
