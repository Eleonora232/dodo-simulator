```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.data_manager import TraitFactory
from src.mainland import Mainland

# 1. Initialize the Factory
factory = TraitFactory(
    bird_path="data/BirdFuncDat.txt",
    mammal_path="data/MamFuncDat.txt",
    reptile_path="data/ReptTraits dataset v1-1.xlsx"
)

# 2. Pull species from a specific Mainland (e.g., Italy)
mainland = Mainland(country_code='IT', factory=factory)
pool_names = mainland.species_pool # This is the list from GBIF

# 3. Build a "Mainland Pool DataFrame" for Analysis
records = []
for name in pool_names:
    traits = factory.get_organism_data(name)
    if traits:
        traits['name'] = name
        # Split name for Genus analysis
        traits['genus'] = name.split()[0] 
        records.append(traits)

df_pool = pd.DataFrame(records)
```
